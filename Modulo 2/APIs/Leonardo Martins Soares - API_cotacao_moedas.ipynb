#### Pegar a cota√ß√£o dos √∫ltimos 30 dias do d√≥lar

```python
cotacao30_dias = requests.get('https://economia.awesomeapi.com.br/json/daily/USD-BRL/30')
cotacao30_dic = cotacao30_dias.json()


#usando o for
for item in cotacao30_dic:
    timestamp = int(item['timestamp'])
    item['timestamp'] = time.strftime('%d/%m/%Y', time.localtime(timestamp))
    print(f"No dia {item['timestamp']} a cota√ß√£o do dolar foi de R${float(item['bid']):,.2f}")
```

#### Pegar as cota√ß√µes do BitCoin de Jan/20 a Out/20
