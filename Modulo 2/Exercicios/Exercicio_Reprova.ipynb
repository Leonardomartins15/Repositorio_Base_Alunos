### Minha Média

```python
### Pegar a nota da prova teorica
### Pedir pro usuario
### Pedir pro usuario a nota da prova pratica
### Calcule a media
### Como eu calculo a média ?? se você não sabe ja reprovou
### Se a média for maior ou igual a 7, print ("aprovado")
### Se não, print ("reprovado")


prova_teorica = 10/10
nota_teorica = float(input(" Qual foi sua nota da prova teorica ?"))
nota_pratica = float(input(" Qual foi sua nota da prova pratica ?"))
total = nota_teorica + nota_pratica
media = total / 2

if media >= 7 :
    print("Aprovado")
else:
    print("Reprovado")
```
